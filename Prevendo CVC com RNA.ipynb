```python
import pandas as pd
import numpy as np
import yfinance as yf

import pandas_ta as ta

import matplotlib.pyplot as plt
```

# Objetivo

- Prever se o próximo fechamento será abaixo ou acima do fechamento atual

# Metodologia Proposta
- Coletar dados
- Suavização Exponencial
- Extração e escolha das melhores features
- Treinamento do modelo

## 1 Coletando dados

```python
data = yf.download('CVCB3.SA', start='2001-01-01', end='2021-12-11')
```

### 1.1 Verificando por valores NA - Missing Values

```python
print(data.isna().sum())
```

## 2 Pré-processamento dos dados - Suavização exponencial

```python
suavizacao = 7
data['Open_S'] = ta.ema(data['Open'], suavizacao)
data['High_S'] = ta.ema(data['High'], suavizacao)
data['Low_S'] = ta.ema(data['Low'], suavizacao)
data['Close_S'] = ta.ema(data['Adj Close'], suavizacao)
data['Volume_S'] = ta.ema(data['Volume'], suavizacao)
```

## 3 Extração e escolha das melhores features

### 3.1 Gerando lags a partir dos fechamentos

```python
for n in np.arange(2, 11, 1):
    data['close_lag_' + str(n)] = data['Close_S'].pct_change(-n)
```

### 3.2 Obtendo indicadores técnicos

```python
# Criando indicadores técnicos
data['obv'] = ta.obv(data['Close_S'], data['Volume_S'])
data['obv_ema13'] = ta.ema(data['obv'], 13)
data['obv_ema21'] = ta.ema(data['obv'], 21)

data['EMA13'] = ta.ema(data['Close_S'], 13)
data['EMA26'] = ta.ema(data['Close_S'], 26)

data['RSI2'] = ta.rsi(data['Close_S'], 2)
data['RSI4'] = ta.rsi(data['Close_S'], 4)

data['Slope13'] = ta.slope(data['Close_S'], 13)
data['Slope21'] = ta.slope(data['Close_S'], 21)

data['roc2'] = ta.roc(data['Close_S'], 2)
data['roc13'] = ta.roc(data['Close_S'], 13)
data['roc21'] = ta.roc(data['Close_S'], 21)

data['linreg15'] = ta.linreg(data['Close_S'], 15)
data['linreg21'] = ta.linreg(data['Close_S'], 21)

data['decaylinear9'] = ta.decay(data['Close_S'], length=9, mode='linear')
data['decaylinear15'] = ta.decay(data['Close_S'], length=15, mode='linear')
data['decaylinear21'] = ta.decay(data['Close_S'], length=21, mode='linear')

data[['zg9', 'sg9', 'xg9', 'atr9']] = ta.aberration(data['High_S'], data['Low_S'], data['Close_S'], length=9)
data[['zg15', 'sg15', 'xg15', 'atr15']] = ta.aberration(data['High_S'], data['Low_S'], data['Close_S'], length=15)
data[['zg21', 'sg21', 'xg21', 'atr21']] = ta.aberration(data['High_S'], data['Low_S'], data['Close_S'], length=21)
```

### 3.3 Gerando mais lags a partir dos indicadores técnicos

```python
for n in np.arange(2, 11, 1):
    data['lag_ema13_' + str(n)] = data['EMA13'].pct_change(-n)
    data['lag_ema21_' + str(n)] = data['EMA26'].pct_change(-n)
```

### 3.4 Removendo os valores faltantes

```python
data = data.dropna()

data = data.drop(columns=['Close'], axis=1)
data = data.reset_index()
data.tail()
```

### 3.5 Gerando mais features através da biblioteca 'featuretools'

```python
import featuretools as ft
```

```python
es = ft.EntitySet('estudo_cvc_fechamento')
es = es.add_dataframe(dataframe_name='cvc', dataframe = data, index = 'Date')
es
```

```python
agg_primitives =  ['max', 'mean', 'median', 'min', 'mode', 'std']
trans_primitives = ['add_numeric', 'divide_numeric',
                    'modulo_numeric', 'multiply_numeric',
                   'subtract_numeric', 'diff']

# Deep feature synthesis 
feature_matrix, feature_names = ft.dfs(entityset=es, target_dataframe_name='cvc',
                       ignore_columns={'cvc': ['Returns']},
                       agg_primitives = agg_primitives,
                       trans_primitives = trans_primitives,
                       max_depth = 1, n_jobs = -1,)
```

```python
# Gerando retorno e removendo valores faltantes
feature_matrix['Returns'] = feature_matrix['Adj Close'].pct_change().shift(-1)
feature_matrix = feature_matrix.dropna()
feature_matrix
```

```python
# Salvando o dataframe
#feature_matrix.to_csv('feature_matrix_cvc_2.csv', index = True)
```

### 3.6 Escolhendo as melhores features

#### 3.6.1 Algoritmo de Classificação: Cálculo das Classes

```python
df = feature_matrix.copy()

df['Target'] = np.where(df['Returns'] > 0, 1, 0)
df = df.drop(columns=['Returns'], axis=1)
df.dropna(inplace=True)
df
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
# Dividindo a base
x = df.drop(columns=['Target'], axis=1)
y = df.iloc[:, -1].values
```

```python
k_best_features = SelectKBest(k='all',)
k_best_features.fit_transform(x, y)
k_best_features_scores = k_best_features.scores_
```

```python
features_list = x.columns.values

raw_pairs = zip(features_list[1:], k_best_features_scores)
ordered_pairs = list(reversed(sorted(raw_pairs, key=lambda x: x[1])))

k_best_features_final = dict(ordered_pairs[:30])
best_features = k_best_features_final.keys()
print ('')
print ("Melhores features:")
print (k_best_features_final)
```

#### 3.6.2 Listando as melhores 30 features

```python
list(best_features)
```

```python
features = df.loc[:,list(best_features)]
features['Target'] = df['Target']
features
```

#### 3.6.3 Observando a correlação entre as features e nossa classe

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr = features.corr(method='pearson')
plt.figure(figsize=(15, 15))
sns.heatmap(corr, annot=True, fmt='.2f', )
```

Removendo as features com baixa correlação com ao Target

```python
features_selecionadas = ['close_lag_2 - lag_ema13_3',
 'close_lag_2 * Low_S',
 'Adj Close * close_lag_3',
 'close_lag_2 * Open_S',
 'close_lag_2 * High_S',
 'close_lag_2 - lag_ema21_3',
 'close_lag_2 * Open',
 'close_lag_2 * RSI2',
 'close_lag_2 * Low',
 'close_lag_2 * EMA13',
 'close_lag_2 * decaylinear21',
 'close_lag_2 * decaylinear9',
 'close_lag_2 * linreg21',
 'close_lag_2 % atr9',
 'close_lag_2 % atr21',
 'RSI4 % close_lag_3',
 'close_lag_2 * xg15',
 'close_lag_3 * Close_S',
 'close_lag_2 * zg15',
 'close_lag_2 - lag_ema21_4',
 'close_lag_3 * Low_S',
 'close_lag_2 * EMA26',
 'close_lag_3 - lag_ema13_3',
 'close_lag_2 * sg21',
 'close_lag_2 * zg21',
 'Adj Close * close_lag_4',
 'close_lag_2 * xg21', 'Target']

features = features[features_selecionadas]
features
```

## 4 Treinando o modelo

### 4.1 Dividindo a base entre treinamento e teste

```python
from sklearn.model_selection import train_test_split
dias = 1 # numero de dias em que se irá prever alta ou queda

X_train,X_test,y_train,y_test=train_test_split(df[features_selecionadas].head(df.shape[0]-dias),
                                               df['Target'].head(df.shape[0]-dias),
                                               test_size= 0.2 ,shuffle=False)
```

### 4.2 Scaling

```python
from sklearn.preprocessing import QuantileTransformer #https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.QuantileTransformer.html

scaler= QuantileTransformer()

X = X_train.values
Y = y_train.values

scaler.fit(X)

X = scaler.transform(X)

X2 = scaler.transform(X_test.values)
Y2 = y_test.values
```

### 4.3 Construindo a rede neural

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout


classifier= Sequential()

# 3 camadas escondidas de 256

#CAMADA DE ENTRADA 
classifier.add(Dense(256, activation='relu', input_dim = X.shape[1]))
classifier.add(Dropout(0.2))

###########################
classifier.add(Dense(128, activation = 'relu'))
classifier.add(Dropout(0.2))

classifier.add(Dense(64, activation = 'relu'))                       
classifier.add(Dropout(0.2))

classifier.add(Dense(32, activation = 'relu'))                       
classifier.add(Dropout(0.2))
#############################

# CAMADA DE SAIDA 
classifier.add(Dense(1, activation = 'sigmoid'))
```

```python
classifier.summary()
```

```python
adam = tensorflow.keras.optimizers.Adam(learning_rate = 0.001)
classifier.compile(optimizer = adam, loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
history = classifier.fit(X ,Y , epochs = 50, batch_size = 32, validation_data = (X2,Y2))
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train','test'], loc='upper right')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Acc')
plt.ylabel('Acc')
plt.xlabel('Epoch')
plt.legend(['Train','test'], loc='upper right')
plt.show()
```

```python
predictions1= classifier.predict(X2)
predictions1 = np.round(predictions1)
```

### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(Y2, predictions1))
```

### Matriz de Confusão

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y2, predictions1)
```
